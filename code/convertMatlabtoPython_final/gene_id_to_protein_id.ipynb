```python
#将geneID转为Proteomics对应的UNIPROTID
path = "F:\python\Bayesian_python"
os.chdir(path)
# 定义函数查找匹配的行
def find_matching_rows(gene_id):
    return df2[df2['gene_name_list'].apply(lambda x: gene_id in x)]

df3 = pd.read_csv('gene_list.txt')
df2 = pd.read_excel('data/uniprotkb_Scer.xlsx', sheet_name= "Sheet0") #蛋白质组学中中找到的gene和protein id
# 对gene name列进行分割
df2['gene_name_list'] = df2['Gene Names'].str.split()
# 对每个Gene_id应用上面的函数，并合并匹配的行
merged_dfs = [find_matching_rows(gene_id) for gene_id in df1['Gene_id']]

merged_data = []

for _, row in df3.iterrows():
    matching_rows = df2[df2['gene_name_list'].apply(lambda x: row['Gene_id'] in x)]
    for _, match_row in matching_rows.iterrows():
        merged_data.append({
            #'No': row['No'],
            'Gene_id': row['Gene_id'],
            #'Species': row['Species'],
            #'Gene': row['Gene'],
            #'Gene_name': row['Gene_name'],
            'Entry': match_row['Entry'],
            'Reviewed': match_row['Reviewed'],
            'Entry Name': match_row['Entry Name'],
            'Protein names': match_row['Protein names'],
            'Gene Names': match_row['Gene Names'],
            'Organism': match_row['Organism'],
            'Length': match_row['Length']
        })

# 转换为DataFrame
final_merged_df = pd.DataFrame(merged_data)
# 保存为新的CSV文件
final_merged_filename = 'data/gene_to_protein.csv'
final_merged_df.to_csv(final_merged_filename, index=False)
```
